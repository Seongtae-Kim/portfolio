## 3. 파일 쓰기

```python
# 기본적으로 파일을 다룰때는 close()함수를 습관화하는 것이 좋다.
f = open("./나의 텍스트.txt", "w")
f.write("안녕하세요. 김성태입니다.")
f.close()
```

## 4. with

```python
# with를 이용할 경우 파일 관련 코드들은 들여쓰기 처리
# with 범위를 벗어날 경우 파일이 자동으로 close()된다.

with open("./나의 텍스트.txt", "w") as f:
    f.write("안녕하세요!")
```

## 8. 파일 함수들

### mode

```python
with open("./세종대왕.txt", "r") as f:
    print(f.mode)
```

### tell()

```python
with open("./세종대왕.txt", "r") as f:
    print(f.tell())
```

### read()

```python
with open("./세종대왕.txt", "r") as f:
    print(f.read())
```

```python
with open("./세종대왕.txt", "r") as f:
    print(f.read(20))
```

```python
with open("./세종대왕.txt", "r") as f:
    s = 50
    text = f.read(s)
    
    while len(text) > 0:
        print(text, end="[구분]")
        text = f.read(s)
```

### seek()

```python
with open("./세종대왕.txt", "r") as f:
    f.seek(20)
    print(f.read(15))
```

```python
with open("./나의 텍스트.txt", "w") as f:
    f.write("환영")
    f.seek(0)
    f.write("수")
```

### readline()

```python
with open("./세종대왕.txt", "r") as f:
    print(f.readline())
```

```python
with open("./세종대왕.txt", "r") as f:
    print("<첫번째 문장>:", f.readline())
    print("<두번째 문장>:", f.readline())
```

```python
with open("./세종대왕.txt", "r") as f:
    i=1
    for l in f:
        print("<", i, "번째 문장>: ", l)
        i+=1
```

### readlines()

```python
with open("./세종대왕.txt", "r") as f:
    print(f.readlines())
```

### writelines()

```python
with open("./나의 텍스트.txt", "w") as f:
    lines = ["첫번째 문장\n", "두번째 문장\n", "세번째 문장\n"]
    f.writelines(lines)
```

## 9. 읽기 쓰기 모드

```python
with open("./나의 텍스트.txt", "r") as f:
    print(f.read())

with open("./나의 텍스트.txt", "w+") as f:
    print(f.tell())
    f.write("새로운 문장!")
    f.seek(0)
    print(f.read())
```

```python
with open("./나의 텍스트.txt", "r") as f:
    print(f.read())

with open("./나의 텍스트.txt", "r+") as f:
    print(f.tell())
    f.write("굉장한")
    f.seek(0)
    print(f.read())
```

```python
with open("./세종대왕.txt", "a+") as f:
    print(f.tell())
    f.write("\n세종대왕은 1450년 음력 2월 17일, 영응대군 집 동별궁에서 승하하였다.")
    f.seek(0)
    print(f.read())
```

## 10. 파일 사본 만들기

### 텍스트 사본 만들기

```python
with open("./세종대왕.txt", "r") as rf:
    with open("./세종대왕_사본.txt", "w") as wf:
        lines = rf.readlines()
        wf.writelines(lines)
```

### 이미지 사본 만들기

```python
with open("./puppy.jpeg", "rb") as rf:
    with open("./puppy_copied.jpeg", "wb") as wf:
        lines = rf.readlines()
        wf.writelines(lines)
```
