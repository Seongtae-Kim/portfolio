```python
def preprocess(series):
    result = set()
    import re, math
    for ele in series:
        if isinstance(ele, str):
            ele = re.sub("(\d+%|\[[^\[\]]+\]|(주)|㈜)", "", ele)
            candids = re.split("[,(·).→~/:\* ∙ㆍ]|\d+\.", ele)
            for candid in candids:
                if not re.search("\d+", candid):
                    result.add(re.sub("\n", "", candid).strip())
        else:
            if not math.isnan(ele):
                continue
    return result

def preprocess_df(df, quiet=False):
    mains = df["주성분"]
    additions = df["첨가제"]
    #product = df["제품명"]
    #company = df["업체명"]
    
    result1 = preprocess(mains)
    result2 = preprocess(additions)
    #result3 = preprocess(product)
    #result4 = preprocess(company)
    
    result = result1.union(result2)
    #result = result.union(result3)
    #result = result.union(result4)
    
    if not quiet:
        print("{} names added".format(len(result)))
    return result
```

```python
from tqdm.notebook import tqdm
result = set()
for path in tqdm(paths):
    df = pd.read_excel(path)
    r = preprocess_df(df, quiet=True)
    result = result.union(r)

print(len(result))
```

```python
import re
result = {re.sub("\[\]\(\)\{\}", "", r) for r in result if re.search("[가-힣]", r)}
print(len(result))
```

# Adding words to Mecab

```python
def check_coda(word):
    import hgtk
    assert isinstance(word, str)
    try:
        r = hgtk.letter.decompose(word[-1])
    except Exception:
        return False
    
    if re.search("[ㄱ-ㅎ]", r[-1]):
        return True
    else:
        return False
```

```python
df = pd.DataFrame(columns=["단어","0","0","0","품사태그","의미분류","종성유무","읽기/타입","첫번째 품사","마지막 품사","원형"])
df
```

```python
from tqdm.notebook import tqdm
for i, word in tqdm(enumerate(result)):
    b = check_coda(word)    
    if b:
        df.loc[i] = [word.strip(), 0, 0, 0, "NNP", "*", "T", "*", "*", "*", "*"]
    else:
        df.loc[i] = [word.strip(), 0, 0, 0, "NNP", "*", "F", "*", "*", "*", "*"]
```

```python
def preprocess_crawled(string):
    import re
    string = re.sub("-","", string)
    string = re.sub("\[약\] .+","", string)
    string = re.sub("\[동\] .+","", string)
    string = re.sub("\[원\] .+","", string)
    string = re.sub("\<속\>","", string)
    string = re.sub("\[각주\].*","", string)
    string = re.sub("\([^\(\)]+\)","", string)
    
    strings = re.split("\d+\.|,| |∙|ㆍ", string)
    strings = [re.sub("[a-zA-Z]+", "", s).strip() for s in strings if re.search("[가-힣]+", s)]
    return strings
```

```python
import re
tmp = set()
for en, ko in terms.items():
    for w in preprocess_crawled(ko):
        tmp.add(w)
```
